```python
import numpy as np
```

```python
# Générons de manière aléaoire un jet de données

np.random.seed(0)
X = 2 * np.random.rand(100, 1)

# La vraie fonctions est 3x + 5 avec un peu de bruit gaussien
y = 5 + 3 * X + np.random.randn(100, 1)
```

```python
# Initialisation des paramètres aléatoires

w = np.random.randn(1)
b = np.random.randn(1)
```

```python
# Les Hyperparamètres

learning_rate = 0.1
epochs = 900
N = len(X)
```

# Comprehension du Gradient Descent

Pour implémenter cela avec NumPy, nous avons deux approches principales :
* **La descente de gradient** (l'approche itérative, fondamentale en Machine Learning).
* **L'équation normale** (l'approche analytique, directe par l'algèbre linéaire).

On a juste besoin de la compréhension du gradient descent car ce qui est la base de ce qu'on fera par après.

## 1. La Descente de Gradient (Itérative)

L'objectif de la régression linéaire simple est de trouver la droite d'équation $y = wx + b$ qui minimise l'Erreur Quadratique Moyenne (MSE).

La fonction de coût (MSE) se définit par :

$$MSE = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$


```python
for epoch in range(epochs):
    # Forward pass (prédiction)
    y_pred = X.dot(w) + b

    # Calcul de l'erreur
    error = y_pred - y
    
    # Calcul du gradient
    dw = (2/N) * X.T.dot(error)
    db = (2/N) * np.sum(error)

    # Mise à jour des paramètres
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Affichons la progression

    if epoch % 100 == 0:
        loss = np.mean(error ** 2) # mse
        print(f"Epoch {epoch}, Loss: {loss}, w: {w}, b: {b}\n")

print(f"Gradient Descent terminé. Paramètres finaux: w: {w}, b: {b}")
```
